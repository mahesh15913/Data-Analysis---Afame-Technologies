```python
ax = sns.countplot(data=data,x = 'OverTime', hue='Attrition')
for p in ax.patches:
    height = p.get_height()
    ax.annotate('{:1.0f}'.format(height), (p.get_x() + p.get_width() / 2., height + 3), ha='center')

plt.show()
```
